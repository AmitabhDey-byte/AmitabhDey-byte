```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test (1).csv')

train.head()
train.info()
```

```python
TARGETS = ['target_a', 'target_b']

features = [c for c in train.columns if c not in ['id'] + TARGETS]

X = train[features].copy()
X_test = test[features].copy()
```

```python
def create_features(df):
    d = df.copy()

    d['row_mean'] = d.mean(axis=1)
    d['row_std'] = d.std(axis=1)
    d['row_min'] = d.min(axis=1)
    d['row_max'] = d.max(axis=1)
    d['row_median'] = d.median(axis=1)
    d['row_skew'] = d.skew(axis=1)
    d['row_kurt'] = d.kurt(axis=1)

    d['missing_count'] = d.isnull().sum(axis=1)

    d['row_q25'] = d.quantile(0.25, axis=1)
    d['row_q75'] = d.quantile(0.75, axis=1)

    return d
```

```python
X = create_features(X)
X_test = create_features(X_test)
```

```python
kf = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
```

```python
oof_lgb = np.zeros((len(train), 2))
pred_lgb = np.zeros((len(test), 2))

oof_cat = np.zeros((len(train), 2))
pred_cat = np.zeros((len(test), 2))
```

```python
for target_idx, target in enumerate(TARGETS):

    y = train[target]

    test_preds_lgb = []
    test_preds_cat = []

    for fold, (train_idx, valid_idx) in enumerate(kf.split(X)):

        X_train = X.iloc[train_idx]
        X_valid = X.iloc[valid_idx]

        y_train = y.iloc[train_idx]
        y_valid = y.iloc[valid_idx]

        model_lgb = LGBMRegressor(
            device='gpu',
            n_estimators=2000,
            learning_rate=0.07,
            num_leaves=40,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42
        )

        model_lgb.fit(
            X_train,
            y_train,
            eval_set=[(X_valid, y_valid)],
            eval_metric='rmse'
        )

        valid_pred_lgb = model_lgb.predict(X_valid)
        test_pred_lgb = model_lgb.predict(X_test)

        oof_lgb[valid_idx, target_idx] = valid_pred_lgb

        test_preds_lgb.append(test_pred_lgb)

        model_cat = CatBoostRegressor(
            task_type='GPU',
            iterations=2000,
            learning_rate=0.056,
            depth=7,
            loss_function='RMSE',
            eval_metric='RMSE',
            random_seed=42,
            verbose=0
        )

        model_cat.fit(
            X_train,
            y_train,
            eval_set=(X_valid, y_valid),
            use_best_model=True
        )

        valid_pred_cat = model_cat.predict(X_valid)
        test_pred_cat = model_cat.predict(X_test)

        oof_cat[valid_idx, target_idx] = valid_pred_cat

        test_preds_cat.append(test_pred_cat)

    pred_lgb[:, target_idx] = np.mean(test_preds_lgb, axis=0)
    pred_cat[:, target_idx] = np.mean(test_preds_cat, axis=0)
```

```python
rmse_lgb_a = np.sqrt(
    mean_squared_error(
        train['target_a'],
        oof_lgb[:, 0]
    )
)

rmse_lgb_b = np.sqrt(
    mean_squared_error(
        train['target_b'],
        oof_lgb[:, 1]
    )
)

rmse_cat_a = np.sqrt(
    mean_squared_error(
        train['target_a'],
        oof_cat[:, 0]
    )
)

rmse_cat_b = np.sqrt(
    mean_squared_error(
        train['target_b'],
        oof_cat[:, 1]
    )
)

print('LGB TARGET A RMSE:', rmse_lgb_a)
print('LGB TARGET B RMSE:', rmse_lgb_b)

print('CAT TARGET A RMSE:', rmse_cat_a)
print('CAT TARGET B RMSE:', rmse_cat_b)
```

```python
final_preds = (
    1 * pred_lgb
)
```

```python
submission = pd.DataFrame({
    'id': test['id'],
    'target_a': final_preds[:, 0],
    'target_b': final_preds[:, 1]
})

submission.to_csv(
    'submission4.csv',
    index=False
)

submission.head()
```
